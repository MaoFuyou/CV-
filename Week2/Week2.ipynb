```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('lenna.jpg',1)
```

```python
img.shape
```

```python
def my_show(img):
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
my_show(img)
```

```python
#Gaussian Kernel
g_img = cv2.GaussianBlur(img,(11,11),40)
#
plt.subplot(121)
my_show(img)
plt.subplot(122)
my_show(g_img)
```

```python
kernel_1d = cv2.getGaussianKernel(11,2)
```

```python
g1_img = cv2.sepFilter2D(img,-1,kernel_1d,kernel_1d)
```

```python
my_show(g1_img)
```

Laplacian

```python
kernel = np.array([[0,1,0],[1,-4,1],[0,1,0]])
kernel
```

```python
#
lap_img = cv2.filter2D(img,-1,kernel)
```

```python
my_show(lap_img)
```

```python
kernel_strong = np.array([[1,1,1],[1,-8,1],[1,1,1]])
kernel_strong
lap_img_strong = cv2.filter2D(img,-1,kernel_strong)
my_show(lap_img_strong)
```

```python
kernel_strong_rui = np.array([[1,1,1],[1,-7,1],[1,1,1]])
kernel_strong_rui
lap_img_strong_rui = cv2.filter2D(img,-1,kernel_strong_rui)
my_show(lap_img_strong_rui)
```

Sobel#

```python
kernel_strong_sobel = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
kernel_strong_sobel
lap_img_strong_sobel = cv2.filter2D(img,-1,kernel_strong_sobel)
my_show(lap_img_strong_sobel)
```

Medianblur

```python
md_img = cv2.medianBlur(img,7)
my_show(md_img)
```

```python
noise_img = cv2.imread('noisy_lenna.jpg',1)
my_show(noise_img)
```

```python
md_img_noise = cv2.medianBlur(noise_img,7)
my_show(md_img_noise)
```

Find feature from figure!
higher/powerful information!

```python
img_gray = np.float32(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY))
```

```python
def my_show_gray(img):
    plt.imshow(img,cmap = 'gray')
```

```python
my_show_gray(img_gray)
```

```python
img_harris = cv2.cornerHarris(img_gray,2,3,0.05)
```

```python
my_show_gray(img_harris)
```

```python
threshold = np.max(img_harris)*0.02
```

```python
img = cv2.imread('lenna.jpg',1)
img[img_harris>threshold] = [0,0,255]#red!!!
```

```python
my_show(img)
```
