```python
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
img_hsv
```

```python
h , s , v  = cv2.split(img_hsv)
```

```python
s += np.clip(s*1.0 + 100,0,255).astype("uint8")
final_hsv = cv2.merge((h,s,v))
```

```python
img_s = cv2.cvtColor(final_hsv,cv2.COLOR_HSV2RGB)
```

```python
plt.figure(dpi = 100)
plt.imshow(img)
plt.axis('off')
```

```python
plt.figure(dpi = 100)
plt.imshow(img_s)
plt.axis('off')
```

```python
immg = cv2.imread('./figures/edge detection.PNG')
```

```python
immg = cv2.cvtColor(immg,cv2.COLOR_BGR2GRAY)

plt.figure(dpi = 200)
plt.imshow(immg, cmap= "gray")
plt.axis("off")
```

```python
laplacian = cv2.Laplacian(immg,cv2.CV_64F, ksize= 5)
sobelx = cv2.Sobel(immg,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(immg,cv2.CV_64F,1,0,ksize=5)

```

```python
plt.figure(dpi = 400)
plt.subplot(2,2,1)
plt.imshow(immg,cmap ='gray')
plt.title('Original')
plt.axis('off')
plt.subplot(2,2,2)
plt.imshow(laplacian,cmap ='gray')
plt.title('laplacian')
plt.axis('off')
plt.subplot(2,2,3)
plt.imshow(sobelx,cmap ='gray')
plt.title('sobelx')
plt.axis('off')
plt.subplot(2,2,4)
plt.imshow(sobely,cmap ='gray')
plt.title('sobely')
plt.axis('off')
```
